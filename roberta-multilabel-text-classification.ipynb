```python
from simpletransformers.classification import MultiLabelClassificationModel


model = MultiLabelClassificationModel('roberta', 'roberta-base', num_labels=50, args={'train_batch_size':2, 'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 'num_train_epochs': 3, 'max_seq_length': 512})
```

```python
model.train_model(train_df)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```
